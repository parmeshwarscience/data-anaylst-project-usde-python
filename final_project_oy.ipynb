## 1.line plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['title'], df['imdb_rating'], marker='o')
plt.title("imdb_rating and title")
plt.xlabel('title')
plt.ylabel('imdb_rating')
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()
```

2.BARPLOT

```python
df_top = df.head(10)

plt.figure(figsize=(10,5))
plt.bar(df_top['title'], df_top['imdb_rating'],color='r')
plt.title("imdb_rating and title (Top 10)")
plt.xlabel('title')
plt.ylabel('imdb_rating')
plt.xticks(rotation=90)
plt.show()
```

3.HISTOGRAM PLOT

```python


plt.figure(figsize=(10,5))
plt.hist(df['imdb_rating'], bins=10)  
plt.title("Distribution of IMDb Ratings")
plt.xlabel('imdb_rating')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()
```

## 4scateer plot

```python
plt.figure(figsize=(10,5))
plt.scatter(df['title'], df['imdb_rating'],color='y')
plt.title("imdb_rating vs title")
plt.xlabel('title')
plt.ylabel('imdb_rating')
plt.xticks(rotation=90)
plt.show()
```

```python
top_shows = df[['title','imdb_rating']].sort_values(by='imdb_rating', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_shows['title'], top_shows['imdb_rating'], color='green')
plt.title("Top 10 Shows by IMDb Rating")
plt.xlabel("Title")
plt.ylabel("IMDb Rating")
plt.xticks(rotation=90)
plt.show()
```
